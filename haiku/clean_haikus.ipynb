```python
import sys, os, re
import fileinput

input_dir = "haiku_files"
filenames = os.listdir(input_dir)
input_filenames = [os.path.join(input_dir, f) for f in filenames if f.endswith(".txt")]
#input_filenames = ["haiku_files/input4.txt"]

print(input_filenames)

output_filename = "../data/input.txt"
```

```python
roman_numbers = "IVXDC"
sign_chars = "[]()"

output_file = open(output_filename, 'w')

n_haikus = 0

for input_filename in input_filenames:
    with open(input_filename) as input_file:
        
        count = 0
        temp=""
        
        for line in input_file:
            if line.strip() and not all(c in roman_numbers for c in line) or any(c in sign_chars for c in line):
                temp += line.lstrip()
                count += 1
            else:
                if count == 3:
                    if not re.search(r'[\d\(\)]', temp):
                        output_file.write(temp + "\n")   
                        n_haikus +=1
                temp = ""
                count = 0
    
output_file.close()

print("haikus:", n_haikus, "\n")

#with open(output_filename) as output_file:
    #print(output_file.read())
    

with fileinput.FileInput(output_filename, inplace=True, backup='.bak') as file:
    for line in file:
        line=line.lower()
        line=re.sub(r'--', ',', line)
        line=re.sub(r'\n\n', '\n\nxxxxx ', line)
        print(line, end='')
        
with open(output_filename) as output_file:
    content = output_file.read()
    
content=content.lower()
content=re.sub(r'--', ',', content)
content=re.sub(r'\n', ' xxx ', content)
content=re.sub(r'xxx  xxx', '\nxxxxxx', content)
    
output_file = open(output_filename, 'w')
output_file.write(content) 
output_file.close()
    
with open(output_filename) as output_file:
    print(output_file.read())
    
```
